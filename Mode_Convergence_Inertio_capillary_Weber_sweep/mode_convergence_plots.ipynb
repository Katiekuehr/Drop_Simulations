```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# File path
file_path = '/Users/katharinakuehr/Drop_Simulations/Mode_Convergence_Inertio_capillary_Weber_sweep/mode_convergence.csv'

# Load the CSV data into a pandas DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to confirm it's loaded correctly
df.head()
```

```python
df_20 = df[df["Modes"] == 20]
df_40 = df[df["Modes"] == 40]
df_60 = df[df["Modes"] == 60]
df_80 = df[df["Modes"] == 80]
df_80 = df_80.drop(df_80.index[-1])
df_100 = df[df["Modes"] == 100]
```

```python
cmap = plt.get_cmap('viridis')

# Get specific colors from the colormap (values between 0 and 1)
colors = [cmap(i) for i in np.linspace(0, 1, 5)]
```

```python
import matplotlib.pyplot as plt

# Create a 2x4 grid of subplots
fig, axs = plt.subplots(2, 4, figsize=(16, 6))

# Adjust the layout to prevent overlap
plt.tight_layout()

# Plotting each dataset into its own subplot
# 1. Coefficient of restitution
axs[0, 0].plot(df_20["We"], df_20["Alpha Coefficient of restitution"], color=colors[0], label="20")
axs[0, 0].plot(df_40["We"], df_40["Alpha Coefficient of restitution"], color=colors[1], label="40")
axs[0, 0].plot(df_60["We"], df_60["Alpha Coefficient of restitution"], color=colors[2], label="60")
axs[0, 0].plot(df_80["We"], df_80["Alpha Coefficient of restitution"], color=colors[3], label="80")
axs[0, 0].plot(df_100["We"], df_100["Alpha Coefficient of restitution"], color=colors[4], label="100")
axs[0, 0].set_xlabel("Weber Number")
axs[0, 0].set_ylabel("Model Prediction")
axs[0, 0].set_title("\n Coefficient of Restitution")
axs[0, 0].legend()

# 2. Contact time ND
axs[0, 1].plot(df_20["We"], df_20["Contact time ND"], color=colors[0], label="20")
axs[0, 1].plot(df_40["We"], df_40["Contact time ND"], color=colors[1], label="40")
axs[0, 1].plot(df_60["We"], df_60["Contact time ND"], color=colors[2], label="60")
axs[0, 1].plot(df_80["We"], df_80["Contact time ND"], color=colors[3], label="80")
axs[0, 1].plot(df_100["We"], df_100["Contact time ND"], color=colors[4], label="100")
axs[0, 1].set_xlabel("Weber Number")
axs[0, 1].set_ylabel("Model Prediction")
axs[0, 1].set_title("\n Contact Time ND")
axs[0, 1].legend()

# 3. Min north pole height ND
axs[0, 2].plot(df_20["We"], df_20["Min north pole height ND"], color=colors[0], label="20")
axs[0, 2].plot(df_40["We"], df_40["Min north pole height ND"], color=colors[1], label="40")
axs[0, 2].plot(df_60["We"], df_60["Min north pole height ND"], color=colors[2], label="60")
axs[0, 2].plot(df_80["We"], df_80["Min north pole height ND"], color=colors[3], label="80")
axs[0, 2].plot(df_100["We"], df_100["Min north pole height ND"], color=colors[4], label="100")
axs[0, 2].set_xlabel("Weber Number")
axs[0, 2].set_ylabel("Model Prediction")
axs[0, 2].set_title("\n Min North Pole Height ND")
axs[0, 2].legend()

# 4. Max radius ND
axs[0, 3].plot(df_20["We"], df_20["Max radius ND"], color=colors[0], label="20")
axs[0, 3].plot(df_40["We"], df_40["Max radius ND"], color=colors[1], label="40")
axs[0, 3].plot(df_60["We"], df_60["Max radius ND"], color=colors[2], label="60")
axs[0, 3].plot(df_80["We"], df_80["Max radius ND"], color=colors[3], label="80")
axs[0, 3].plot(df_100["We"], df_100["Max radius ND"], color=colors[4], label="100")
axs[0, 3].set_xlabel("Weber Number")
axs[0, 3].set_ylabel("Model Prediction")
axs[0, 3].set_title("\n Max Radius ND")
axs[0, 3].legend()

# 5. Max contact radius ND
axs[1, 0].plot(df_20["We"], df_20["Max contact radius ND"], color=colors[0], label="20")
axs[1, 0].plot(df_40["We"], df_40["Max contact radius ND"], color=colors[1], label="40")
axs[1, 0].plot(df_60["We"], df_60["Max contact radius ND"], color=colors[2], label="60")
axs[1, 0].plot(df_80["We"], df_80["Max contact radius ND"], color=colors[3], label="80")
axs[1, 0].plot(df_100["We"], df_100["Max contact radius ND"], color=colors[4], label="100")
axs[1, 0].set_xlabel("Weber Number")
axs[1, 0].set_ylabel("Model Prediction")
axs[1, 0].set_title("\n Max Contact Radius ND")
axs[1, 0].legend()

# 6. Max radial projection ND
axs[1, 1].plot(df_20["We"], df_20["Max radial projection ND"], color=colors[0], label="20")
axs[1, 1].plot(df_40["We"], df_40["Max radial projection ND"], color=colors[1], label="40")
axs[1, 1].plot(df_60["We"], df_60["Max radial projection ND"], color=colors[2], label="60")
axs[1, 1].plot(df_80["We"], df_80["Max radial projection ND"], color=colors[3], label="80")
axs[1, 1].plot(df_100["We"], df_100["Max radial projection ND"], color=colors[4], label="100")
axs[1, 1].set_xlabel("Weber Number")
axs[1, 1].set_ylabel("Model Prediction")
axs[1, 1].set_title("\n Max Radial Projection ND")
axs[1, 1].legend()

# 7. Min side height ND
axs[1, 2].plot(df_20["We"], df_20["Min side height ND"], color=colors[0], label="20")
axs[1, 2].plot(df_40["We"], df_40["Min side height ND"], color=colors[1], label="40")
axs[1, 2].plot(df_60["We"], df_60["Min side height ND"], color=colors[2], label="60")
axs[1, 2].plot(df_80["We"], df_80["Min side height ND"], color=colors[3], label="80")
axs[1, 2].plot(df_100["We"], df_100["Min side height ND"], color=colors[4], label="100")
axs[1, 2].set_xlabel("Weber Number")
axs[1, 2].set_ylabel("Model Prediction")
axs[1, 2].set_title("\n Min Side Height ND")
axs[1, 2].legend()

# Leave axs[1, 3] empty
axs[1, 3].axis('off')  # This makes the subplot empty
plt.subplots_adjust(hspace=0.3) 

# Show the plot
plt.show()
```

```python
modes = [df_20, df_40, df_60, df_80, df_100]

coef_res_improvements = []
contact_time_improvements = []
north_pole_h_improvements = []
max_radius_improvements = []
max_contact_radius_improvements = []
max_radial_projectiom_improvements = []
min_side_height_improvements = []

Webers = [0.0001, 0.001, 0.01, 0.2, 0.4, 0.6]
current_df = df_40[df_20["We"] == 0.0001]
next_df = df_60[df_40["We"] == 0.0001]
current_cr = current_df["Max contact radius ND"].values[0]
next_cr = next_df["Max contact radius ND"].values[0]
print(current_cr, next_cr, current_cr-next_cr, (current_cr-next_cr)/current_cr)



for mode in range(len(modes)-1):

    current_mode = modes[mode]
    next_mode = modes[mode + 1]

    store_coef_improvements = []
    store_contact_time_improvements = []
    store_north_pole_improvements = []
    store_max_radius_improvements = []
    store_contact_radius_improvements = []
    store_max_radial_improvements = []
    store_min_side_height_improvements = []

    for Weber in Webers:
        
        if Weber != 0.6:
            current_df = current_mode[current_mode["We"] == Weber]
            following_df = next_mode[next_mode["We"] == Weber]
            
            current_alpha = current_df["Alpha Coefficient of restitution"].values[0] 
            next_alpha = following_df["Alpha Coefficient of restitution"].values[0]
            relative_alpha_dif = np.absolute((current_alpha-next_alpha)/current_alpha)*100
            store_coef_improvements.append(relative_alpha_dif)

            current_cont_time = current_df["Contact time ND"].values[0] 
            next_cont_time = following_df["Contact time ND"].values[0]
            relative_cont_time_dif = np.absolute((current_cont_time-next_cont_time)/current_cont_time)*100
            store_contact_time_improvements.append(relative_cont_time_dif)

            current_north_pole = current_df["Min north pole height ND"].values[0] 
            next_north_pole = following_df["Min north pole height ND"].values[0]
            rel_north_pole_h_dif = np.absolute((current_north_pole-next_north_pole)/current_north_pole)*100
            store_north_pole_improvements.append(rel_north_pole_h_dif)

            current_max_rad = current_df["Max radius ND"].values[0] 
            next_max_rad = following_df["Max radius ND"].values[0]
            rel_max_rad_dif = np.absolute((current_max_rad-next_max_rad)/current_max_rad)*100
            store_max_radius_improvements.append(rel_max_rad_dif)

            current_max_cont_rad = current_df["Max contact radius ND"].values[0] 
            next_max_cont_rad = following_df["Max contact radius ND"].values[0]
            rel_max_cont_rad_dif = np.absolute((current_max_cont_rad-next_max_cont_rad)/current_max_cont_rad)*100
            store_contact_radius_improvements.append(rel_max_cont_rad_dif)

            current_max_rad_proj = current_df["Max radial projection ND"].values[0] 
            next_max_rad_proj = following_df["Max radial projection ND"].values[0]
            rel_max_rad_proj_dif = np.absolute((current_max_rad_proj-next_max_rad_proj)/current_max_rad_proj)*100
            store_max_radial_improvements.append(rel_max_rad_proj_dif)

            current_min_side_h = current_df["Min side height ND"].values[0] 
            next_min_side_h = following_df["Min side height ND"].values[0]
            rel_min_side_h_dif = np.absolute((current_min_side_h-next_min_side_h)/current_min_side_h)*100
            store_min_side_height_improvements.append(rel_min_side_h_dif)

        if Weber == 0.6 and mode == 0:
            current_df = current_mode[current_mode["We"] == Weber]
            following_df = next_mode[next_mode["We"] == Weber]
            
            current_alpha = current_df["Alpha Coefficient of restitution"].values[0] 
            next_alpha = following_df["Alpha Coefficient of restitution"].values[0]
            relative_alpha_dif = np.absolute((current_alpha-next_alpha)/current_alpha)*100
            store_coef_improvements.append(relative_alpha_dif)

            current_cont_time = current_df["Contact time ND"].values[0] 
            next_cont_time = following_df["Contact time ND"].values[0]
            relative_cont_time_dif = np.absolute((current_cont_time-next_cont_time)/current_cont_time)*100
            store_contact_time_improvements.append(relative_cont_time_dif)

            current_north_pole = current_df["Min north pole height ND"].values[0] 
            next_north_pole = following_df["Min north pole height ND"].values[0]
            rel_north_pole_h_dif = np.absolute((current_north_pole-next_north_pole)/current_north_pole)*100
            store_north_pole_improvements.append(rel_north_pole_h_dif)

            current_max_rad = current_df["Max radius ND"].values[0] 
            next_max_rad = following_df["Max radius ND"].values[0]
            rel_max_rad_dif = np.absolute((current_max_rad-next_max_rad)/current_max_rad)*100
            store_max_radius_improvements.append(rel_max_rad_dif)

            current_max_cont_rad = current_df["Max contact radius ND"].values[0] 
            next_max_cont_rad = following_df["Max contact radius ND"].values[0]
            rel_max_cont_rad_dif = np.absolute((current_max_cont_rad-next_max_cont_rad)/current_max_cont_rad)*100
            store_contact_radius_improvements.append(rel_max_cont_rad_dif)

            current_max_rad_proj = current_df["Max radial projection ND"].values[0] 
            next_max_rad_proj = following_df["Max radial projection ND"].values[0]
            rel_max_rad_proj_dif = np.absolute((current_max_rad_proj-next_max_rad_proj)/current_max_rad_proj)*100
            store_max_radial_improvements.append(rel_max_rad_proj_dif)

            current_min_side_h = current_df["Min side height ND"].values[0] 
            next_min_side_h = following_df["Min side height ND"].values[0]
            rel_min_side_h_dif = np.absolute((current_min_side_h-next_min_side_h)/current_min_side_h)*100
            store_min_side_height_improvements.append(rel_min_side_h_dif)


    coef_res_improvements.append(store_coef_improvements)
    contact_time_improvements.append(store_contact_time_improvements)
    north_pole_h_improvements.append(store_north_pole_improvements)
    max_radius_improvements.append(store_max_radius_improvements)
    max_contact_radius_improvements.append(store_contact_radius_improvements)
    max_radial_projectiom_improvements.append(store_max_radial_improvements)
    min_side_height_improvements.append(store_min_side_height_improvements)

print(coef_res_improvements)
```

```python
import matplotlib.pyplot as plt

# Create a 2x4 grid of subplots
fig, axs = plt.subplots(2, 4, figsize=(20, 8))

# Adjust the layout to prevent overlap
plt.tight_layout()
weber_short = df_20["We"].drop(df_20["We"].index[-1])

# Plotting each dataset into its own subplot
# 1. Coefficient of restitution
axs[0, 0].plot(df_20["We"], coef_res_improvements[0], color=colors[0], label="|(M20-M40)|/M20")
axs[0, 0].plot(weber_short, coef_res_improvements[1], color=colors[2], label="|(M40-M60)|/M40")
axs[0, 0].plot(weber_short, coef_res_improvements[2], color=colors[3], label="|(M60-M80)|/M60")
axs[0, 0].plot(weber_short, coef_res_improvements[3], color=colors[4], label="|(M80-M100)|/M80")
axs[0, 0].set_title("\n Coefficient of Restitution")
axs[0, 0].set_xlabel("Weber Number")
axs[0, 0].set_ylabel("Percentage Difference in Prediction")
axs[0, 0].legend()

# 2. Contact time ND
axs[0, 1].plot(df_20["We"], contact_time_improvements[0], color=colors[0], label="|(M20-M40)|/M20")
axs[0, 1].plot(weber_short, contact_time_improvements[1], color=colors[2], label="|(M40-M60)|/M40")
axs[0, 1].plot(weber_short, contact_time_improvements[2], color=colors[3], label="|(M60-M80)|/M60")
axs[0, 1].plot(weber_short, contact_time_improvements[3], color=colors[4], label="|(M80-M100)||/M80")
axs[0, 1].set_title("\n Contact Time ND")
axs[0, 1].set_xlabel("Weber Number")
axs[0, 1].set_ylabel("Absolute Percentage Difference in Prediction")
axs[0, 1].legend()

# 3. Min north pole height ND
axs[0, 2].plot(df_20["We"], north_pole_h_improvements[0], color=colors[0], label="|(M20-M40)|/M20")
axs[0, 2].plot(weber_short, north_pole_h_improvements[1], color=colors[2], label="|(M40-M60)|/M40")
axs[0, 2].plot(weber_short, north_pole_h_improvements[2], color=colors[3], label="|(M60-M80)|/M60")
axs[0, 2].plot(weber_short, north_pole_h_improvements[3], color=colors[4], label="|(M80-M100)|/M80")
axs[0, 2].set_title("\n Min North Pole Height ND")
axs[0, 2].set_xlabel("Weber Number")
axs[0, 2].set_ylabel("Percentage Difference in Prediction")
axs[0, 2].legend()

# 4. Max radius ND
axs[0, 3].plot(df_20["We"], max_radius_improvements[0], color=colors[0], label="|(M20-M40)|/M20")
axs[0, 3].plot(weber_short, max_radius_improvements[1], color=colors[2], label="|(M40-M60)|/M40")
axs[0, 3].plot(weber_short, max_radius_improvements[2], color=colors[3], label="|(M60-M80)|/M60")
axs[0, 3].plot(weber_short, max_radius_improvements[3], color=colors[4], label="|(M80-M100)|/M80")
axs[0, 3].set_ylim(0,1)
axs[0, 3].set_title("\n Max Radius ND")
axs[0, 3].set_xlabel("Weber Number")
axs[0, 3].set_ylabel("Percentage Difference in Prediction")
axs[0, 3].legend()

# 5. Max contact radius ND
axs[1, 0].plot(df_20["We"], max_contact_radius_improvements[0], color=colors[0], label="|(M20-M40)|/M20")
axs[1, 0].plot(weber_short, max_contact_radius_improvements[1], color=colors[2], label="|(M40-M60)|/M40")
axs[1, 0].plot(weber_short, max_contact_radius_improvements[2], color=colors[3], label="|(M60-M80)|/M60")
axs[1, 0].plot(weber_short, max_contact_radius_improvements[3], color=colors[4], label="|(M80-M100)|/M80")
axs[1, 0].set_title("\n Max Contact Radius ND")
axs[1, 0].set_xlabel("Weber Number")
axs[1, 0].set_ylabel("Percentage Difference in Prediction")
axs[1, 0].legend()

# 6. Max radial projection ND
axs[1, 1].plot(df_20["We"], max_radial_projectiom_improvements[0], color=colors[0], label="|(M20-M40)|/M20")
axs[1, 1].plot(weber_short, max_radial_projectiom_improvements[1], color=colors[2], label="|(M40-M60)|/M40")
axs[1, 1].plot(weber_short, max_radial_projectiom_improvements[2], color=colors[3], label="|(M60-M80)|/M60")
axs[1, 1].plot(weber_short, max_radial_projectiom_improvements[3], color=colors[4], label="|(M80-M100)|/M80")
axs[1, 1].set_ylim(0,1)
axs[1, 1].set_title("\n Max Radial Projection ND")
axs[1, 1].set_xlabel("Weber Number")
axs[1, 1].set_ylabel("Percentage Difference in Prediction")
axs[1, 1].legend()

# 7. Min side height ND
axs[1, 2].plot(df_20["We"], min_side_height_improvements[0], color=colors[0], label="|(M20-M40)|/M20")
axs[1, 2].plot(weber_short, min_side_height_improvements[1], color=colors[2], label="|(M40-M60)|/M40")
axs[1, 2].plot(weber_short, min_side_height_improvements[2], color=colors[3], label="|(M60-M80)|/M60")
axs[1, 2].plot(weber_short, min_side_height_improvements[3], color=colors[4], label="|(M80-M100)|/M80")
axs[1, 2].set_ylim(0,1)
axs[1, 2].set_title("\n Min Side Height ND")
axs[1, 2].set_xlabel("Weber Number")
axs[1, 2].set_ylabel("Percentage Difference in Prediction")
axs[1, 2].legend()

# Leave axs[1, 3] empty
axs[1, 3].axis('off')  # This makes the subplot empty
plt.subplots_adjust(hspace=0.3) 

# Show the plot
plt.show()
```
